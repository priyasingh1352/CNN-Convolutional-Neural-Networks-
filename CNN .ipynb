```python
import tensorflow
```

```python
from keras.layers import Dense , Conv2D , MaxPooling2D , Flatten # as it is case sensitive write in same form as it is.
```

```python
from keras.models import Sequential                   
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
cnn = Sequential()
```

```python
cnn.add(Conv2D(32,(3,3),input_shape = (64,64,3),activation = "relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Conv2D(32,(3,3),activation = "relu"))
cnn.add(MaxPooling2D(pool_size=(2,2)))
cnn.add(Flatten())
```

```python
cnn.add(Dense(64,activation = "relu"))
cnn.add(Dense(32,activation = "relu"))
cnn.add(Dense(16,activation = "relu"))
cnn.add(Dense(8,activation = "relu"))
cnn.add(Dense(4,activation = "relu"))
cnn.add(Dense(1,activation = "sigmoid"))
```

```python
cnn.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
       r"C:\Users\Priya singh\OneDrive\Desktop\Extracted Data for DATA SCIENCE\cnn cat dog dataset test and train\training_set\training_set",
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')

test_generator = test_datagen.flow_from_directory(
       r"C:\Users\Priya singh\OneDrive\Desktop\Extracted Data for DATA SCIENCE\cnn cat dog dataset test and train\test_set",
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')

cnn.fit(
        train_generator,
        steps_per_epoch=2000,
        epochs=50,
        validation_data=test_generator)
```
